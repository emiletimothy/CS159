```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib
from matplotlib.patches import Ellipse
import scipy as sc
from scipy import stats
from scipy import ndimage
import sklearn as sk
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import manifold, datasets, decomposition
import pandas as pd
import seaborn as sns
import random
import tensorflow as tf
import torch
import torchvision
import torchvision.transforms as transforms
from scipy.stats import pearsonr
```

```python
url="https://raw.githubusercontent.com/emiletimothy24/CS159/main/data/C__Users_Emile_Desktop_CS159ActualSyntheticData%20(1).csv"
df=pd.read_csv(url)
```

```python
df
```

```python
X = [[0 for i in range(20)] for j in range(6000)]
for i in range(20):
  for j,k in enumerate(list(df[df.columns[i]])):
    X[j][i]=k
Y = list(df['Y'])
X = np.array(X)
```

```python
perplexity=20
print("\n\nPerplexity = ", perplexity)
tsne = manifold.TSNE(n_components=2, init='random', method='barnes_hut', random_state=0, perplexity=perplexity,n_iter=650, verbose=2,learning_rate=1.5)
Yprime = tsne.fit_transform(X[:,0:20])
```

```python
fig = plt.figure(figsize=(8, 6))
ax = plt.axes()
string = "Perplexity = " + str(perplexity)
plt.title(string)
cmap = matplotlib.cm.get_cmap('magma')
normalize = matplotlib.colors.Normalize(vmin=min(Y), vmax=max(Y))
colors = [cmap(normalize(value)) for value in Yprime]
colors = [list(np.ndarray.flatten(np.array(i))) for i in colors]
sc = plt.scatter(Yprime[:,0], Yprime[:,1], c=Y)
cax, _ = matplotlib.colorbar.make_axes(ax)
cbar = matplotlib.colorbar.ColorbarBase(cax, cmap=cmap, norm=normalize)
plt.show()
```

```python
def spectral_density_original(index):
  X_vals = X[300*index:300*(index+1)]
  pca = PCA(n_components=2)
  pca.fit(X_vals)
  X0 = pca.transform(X_vals)
  sample_cov = np.cov(X0.T)
  _, D, _ = np.linalg.svd(sample_cov, full_matrices=False)
  return np.sum(D**2), X0 #spectral density of the PCA

def spectral_density_tsne(index):
  X_vals = Yprime[300*index:300*(index+1)]
  sample_cov = np.cov(X_vals.T)
  _, D, _ = np.linalg.svd(sample_cov, full_matrices=False)
  return np.sum(D**2), X_vals #spectral density of the tSNE
```

```python
orig_fig,x=[],[]
tsne_fig,y=[],[]
for i in range(20):
  ai, xi = spectral_density_original(i)
  orig_fig.append(xi)
  x.append(ai)
  bi, yi = spectral_density_tsne(i)
  tsne_fig.append(yi)
  y.append(bi)
plt.title("Spectral Densities of tSNE vs Data")
plt.xlabel("Spectral Densities of Data from PCA")
plt.ylabel("Spectral Densities of Data from TSNE")
plt.scatter(x,y,color='black')
corr, _ = pearsonr(x,y)
print("Pearson Correlation Coefficient is ", corr)
```

```python
val = sc.stats.spearmanr(x,y)[0]#.correlation
print(val)
```

```python
fig, axs = plt.subplots(20,2)
fig.set_figheight(60)
fig.set_figwidth(10)
for i in range(len(orig_fig)):
  axs[i,0].scatter(orig_fig[i][:,0],orig_fig[i][:,1])
  axs[i,1].scatter(tsne_fig[i][:,0],tsne_fig[i][:,1])
```

Experiment 2: Does distance in tSNE correlate to distance in raw data?

```python
def compute_mean_coordinate_for_original_cluster(index):
  X_vals = X[300*index:300*(index+1)]
  return [np.mean(X_vals[:,i]) for i in range(20)]

def compute_mean_coordinate_for_tsne_cluster(index):
  X_vals = Yprime[300*index:300*(index+1)]
  return [np.mean(X_vals[:,i]) for i in range(2)]

def compute_distances_from_center_original(index):
  mean = compute_mean_coordinate_for_original_cluster(index)
  distance_vectors = X[300*index:300*(index+1)] - mean
  distance = []
  for vec in distance_vectors:
    distance.append(np.sqrt(np.sum(vec*vec)))
  return distance

def compute_distances_from_tsne_cluster(index):
  mean = compute_mean_coordinate_for_tsne_cluster(index)
  distance_vectors = Yprime[300*index:300*(index+1)] - mean
  distance = []
  for vec in distance_vectors:
    distance.append(np.sqrt(np.sum(vec*vec)))
  return distance
```

```python
original_distance = []
tsne_distance = []
for index in range(20):
  original_distance.append(compute_distances_from_center_original(index))
  tsne_distance.append(compute_distances_from_tsne_cluster(index))
```

```python
fig, axs = plt.subplots(10, 2)
fig.set_figheight(60)
fig.set_figwidth(10)
for i in range(10):
  s0 = "Distance plot for class " + str(2*i)
  s1 = "Distance plot for class " + str(2*i+1)
  axs[i,0].set_title(s0)
  axs[i,1].set_title(s1)
  axs[i,0].scatter(original_distance[2*i],tsne_distance[2*i],color='black')
  axs[i,0].set_xlabel("Distance in the synthetic dataset")
  axs[i,0].set_ylabel("Distance in t-SNE")
  axs[i,1].scatter(original_distance[2*i+1],tsne_distance[2*i+1],color='black')
  axs[i,1].set_xlabel("Distance in the synthetic dataset")
  axs[i,1].set_ylabel("Distance in t-SNE")
  ax.set_xlabel('common xlabel')
```

```python
for i in range(20):
  pl = "Class " + str(i) + " = "
  val = sc.stats.spearmanr(original_distance[i],tsne_distance[i])[0]#.correlation
  print(pl + str(val))
```

```python
all_distances_original = []
all_distances_tsne = []
for i,j in zip(original_distance, tsne_distance):
  for xi,yi in zip(i,j):
    all_distances_original.append(xi)
    all_distances_tsne.append(yi)
```
